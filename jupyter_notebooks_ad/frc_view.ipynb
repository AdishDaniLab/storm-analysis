To view the Fourier Ring Correlation result.

```python
import os
import pandas as pd 
import matplotlib.pyplot as plt
import numpy

# Change directory
os.chdir("/home/gayatri/storm/05-22-20_brain")
```

```python
frc_result = 'frc_filt_res.txt'
df = pd.read_csv(frc_result, sep=' ', header=None)
df.columns = ['xvals', 'frc']
df.head()
```

```python
plt.figure(figsize = (8,8))
d6 = plt.scatter(df['xvals'], df['frc'], s=10, alpha=0.5, c='teal')
plt.xlim([0.0, 0.03])
# plt.ylim([0, 40000])
plt.title('FRC Results ')
plt.ylabel('Correlation')
plt.xlabel('Spatial Frequency(nm-1)')
plt.savefig('frc_plot_filtered')
```
