## Comparing synapses to find optimal imaging conditions ##  
Several movies were acquired recently with different imaging conditions. However, all of them produced visually clear synaptic structures. I am trying to compare these synapses qualitatively/quantitatively to figure out which combination of imaging conditions (like shutter sequence, dichroic, frame rate, etc) gives high sampling density. Only those synapses that are in focus are considered for analysis.  


```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy

# Change directory
os.chdir("/home/gayatri/storm-data/synapse_density/20-08-29")
```

```python
m1 = 'm1_synapse_list.txt'
m3 = 'm3_synapse_list.txt'
m4 = 'm4_synapse_list.txt'
m5 = 'm5_synapse_list.txt'
m7 = 'm7_synapse_list.txt'

df1 = pd.read_csv(m1, sep='\t')
df3 = pd.read_csv(m3, sep='\t')
df4 = pd.read_csv(m4, sep='\t')
df5 = pd.read_csv(m5, sep='\t')
df7 = pd.read_csv(m7, sep='\t')
# file_name = mlist.rsplit('.txt',1)[0][:]
```

```python
fig, ax = plt.subplots(figsize = (8,8))
df1['X'][df1['Frame']==1].hist(bins=20,alpha=0.5, density=False)
df3['X'][df3['Frame']==1].hist(bins=20,alpha=0.5, density=False)
df4['X'][df4['Frame']==1].hist(bins=20,alpha=0.5, density=False)
df5['X'][df5['Frame']==2].hist(bins=20,alpha=0.5, density=False)
df7['X'][df7['Frame']==1].hist(bins=20,alpha=0.5, density=False)
# No. of synapses in each molecule list :
# m1 : 1,2,3,4,5
# m3 : 1
# m4 : 1
# m5 : 1,2
# m7 : 1,2,3
plt.ylim([0, 110])
ax.legend(labels = ('70mW, 2+5', '90mW, 5+20', '5+20', '2+10', '1+5'))
```
