## Quantifying image quality (Part 2/2)
For combining data drom 3 csv files (from different frames) into one. 
Concerned data-folder : session16 

```python
import os
import matplotlib.pyplot as plt
import glob
import pandas as pd

os.chdir("D:/gayatri-folder/session16")

# Load images
csv_files = sorted(glob.glob('D:/gayatri-folder/session16/all*.csv'))
```

```python
dataframes = []
for file in csv_files:
    df = pd.read_csv(file)
    dataframes.append(df)
combined = pd.concat(dataframes)
combined.to_csv("snr_analysis.csv", header=['row','col','x','y','peak_ph','avg_ph', 'sum_ph', 'snr1', 'snr2', 'snr3', 'snr4', 'exp'],index=False)
```

```python
plt.figure(figsize = (20,8))
d1 = plt.scatter(combined['exp'], combined['snr4'], color = 'orchid', marker = ".")
plt.ylim(0,5)
plt.ylabel('Signal-to-background')
# plt.xlabel('Experiments')
plt.savefig('snr4_scatter.png')
plt.show()
```
